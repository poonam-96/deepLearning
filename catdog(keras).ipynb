```python
files = os.listdir('data')
```

```python
ffile = files[1:500]
ffile
    
```

```python
X = []
Y = []
for item in tqdm(ffile):
    img = cv2.imread('data/'+ item)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224,224))
    if 'cat' in item:
      Y.append(1)
      X.append(img)
    if 'dog' in item:
      Y.append(0)
      X.append(img)
```

```python
X = np.array(X)
X
```

```python

X = X / 255.0
```

```python
from keras.applications.inception_v3 import InceptionV3
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K

# create the base pre-trained model
base_model = InceptionV3(weights='imagenet', include_top=False)

# add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
# let's add a fully-connected layer
x = Dense(1024, activation='relu')(x)
# and a logistic layer --https://towardsdatascience.com/transfer-learning-from-pre-trained-models-f2393f124751
predictions = Dense(1, activation='sigmoid')(x)

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainablle = False
```

```python
from keras import optimizers
ad = optimizers.Adam(lr=0.1)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=ad,
              metrics=['accuracy'])

model.fit(X, Y, batch_size=8, epochs=2, validation_split=0.2)
```
